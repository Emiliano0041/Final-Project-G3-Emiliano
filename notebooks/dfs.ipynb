```python
import pandas as pd
from glob import glob
import os
```

```python
data_dir_1 = '../data/raw/accidents_causa_conductor'
causa_conductor_files = sorted(glob(os.path.join(data_dir_1, "*.csv")))
data_dir_2 = '../data/raw/accidents_causa_vianant'
accident_files = sorted(glob(os.path.join(data_dir_2, "*.csv")))
data_dir_3 = '../data/raw/accidents_causa_mediata'
causa_mediata_files = sorted(glob(os.path.join(data_dir_3, "*.csv")))
data_dir_4 = '../data/raw/accidents_segun_tipologia'
segun_tipologia_files = sorted(glob(os.path.join(data_dir_4, "*.csv")))
```

```python
df_causa_conductor_2 = pd.concat([pd.read_csv(file) for file in causa_conductor_files ], ignore_index=True)
df_accidents = pd.concat([pd.read_csv(file) for file in accident_files], ignore_index=True)
df_causa_mediata_2 = pd.concat([pd.read_csv(file) for file in causa_mediata_files ], ignore_index=True)
df_segun_tipologia_2 = pd.concat([pd.read_csv(file) for file in segun_tipologia_files ], ignore_index=True)
```

```python
df_causa_mediata_2['Descripcio_causa_mediata'] = df_causa_mediata_2['Descripcio_causa_mediata'].combine_first(df_causa_mediata_2['Descripció_causa_mediata'])
df_causa_mediata_2['Numero_expedient'] = df_causa_mediata_2['Numero_expedient'].combine_first(df_causa_mediata_2["Número_d'expedient"])
```

```python
df_segun_tipologia_2['Numero_expedient'] = df_segun_tipologia_2['Numero_expedient'].combine_first(df_segun_tipologia_2["Codi_expedient"])
df_segun_tipologia_2['Descripcio_tipus_accident'] = df_segun_tipologia_2['Descripcio_tipus_accident'].combine_first(df_segun_tipologia_2['Tipus_accident'])
```

```python
# Extraer las columnas clave como sets
set_accidents = set(df_accidents['Numero_expedient'].dropna())
set_causa = set(df_causa_conductor_2['Numero_expedient'].dropna())

# Obtener la intersección
coincidentes = set_accidents.intersection(set_causa)

# Mostrar los expedientes que coinciden
print(f"Número de expedientes coincidentes: {len(coincidentes)}")
```
