```python
import pandas as pd
from glob import glob
import os
import pickle
from pickle import load
import holidays
```

```python
data_dir = '../data/raw'
# Buscar todos los CSV de la carpeta
cultura_files = sorted(glob(os.path.join(data_dir, "*.csv")))
cultura_files
```

```python
df_eventos = pd.concat([pd.read_csv(file) for file in cultura_files], ignore_index=True)
df_eventos
```

```python
df_eventos.info()
```

```python
df_eventos["DataInici"] = pd.to_datetime(df_eventos["DataInici"], format="%d/%m/%Y")
df_eventos["DataFi"] = pd.to_datetime(df_eventos["DataFi"], format="%d/%m/%Y")
```

```python
filas = []
for _, row in df_eventos.iterrows():
    dias = pd.date_range(start=row["DataInici"], end=row["DataFi"])
    for d in dias:
        filas.append({"Festa_nom": row["Festa"], "fecha": d})
```

```python
df_fiestas = pd.DataFrame(filas)
df_fiestas
```

```python
df_festas = pd.concat([df_eventos, df_fiestas], axis=1)
df_festas
```

```python
df_festas.drop(["Any", "Festa", "DataInici", "DataFi", "Nota"], axis=1, inplace=True)
df_festas
```

```python
festivos_cat = holidays.Spain(years=list(range(2017, 2025)), prov="CT")
df_festas["Festivo"] = df_festas["fecha"].isin(festivos_cat).astype(int)
df_festas
```

```python
df_festas["Festivo"].value_counts()
```

```python
df_festas.to_csv("Df_fiestas.csv", index=False)
```
