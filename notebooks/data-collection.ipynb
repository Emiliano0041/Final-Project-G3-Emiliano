```python
# Ver si hay valores duplicados del numero de expediente
duplicados = df_accidents[df_accidents["Numero_expedient"].duplicated()]
duplicados
```

```python
# Encontrar los expedientes duplicados (que aparecen más de una vez)
duplicados_ids = df_accidents["Numero_expedient"][df_accidents["Numero_expedient"].duplicated(keep=False)]
duplicados_ids
```

```python
# Filtrar todos los registros con los expedientes duplicados
expedientes_duplicados= df_accidents[df_accidents["Numero_expedient"].isin(duplicados_ids)]
expedientes_duplicados.sort_values("Numero_expedient")
```

```python
col_fijas = [c for c in df_accidents.columns if c != 'Descripcio_causa_vianant']

df_accidents = (
    df_accidents
    .groupby("Numero_expedient", as_index=False)
    .agg({
        **{col: 'first' for col in col_fijas if col != "Numero_expedient"},
        'Descripcio_causa_vianant': lambda x: (
            pd.NA if x.isna().all() 
            else ", ".join(sorted(set(v for v in x if pd.notna(v) and str(v).strip() != "")))
        )
    })
)
```

```python
# Comprobamos que ya no hay expedientes duplicados
df_accidents[df_accidents["Numero_expedient"].isin(duplicados_ids)]
```

```python
df_accidents.duplicated().sum()
```

```python
# Actualiza la columna existente Any con los valores de NK_Any cuando Any está vacía (NaN)
df_accidents["Any"] = df_accidents["Any"].combine_first(df_accidents["NK_Any"])

# Eliminar la columna redundante
df_accidents = df_accidents.drop(columns=["NK_Any"])
```

```python
df_accidents = limpiar_clave(df_accidents, "Numero_expedient")

# --- Dataset 1: causa conductor ---
df_causa_conductor = cargar_y_unificar(
    "../data/raw/accidents_causa_conductor",
    ["Numero_expedient", "Número_expedient"],
    ["Descripcio_causa_mediata", "Descripcio_torn"]
)

# --- Dataset 2: causa mediata ---
df_causa_mediata = cargar_y_unificar(
    "../data/raw/accidents_causa_mediata",
    ["Numero_expedient", "Número_d'expedient"],
    ["Descripcio_causa_mediata", "Descripció_causa_mediata"]
)

# --- Dataset 3: tipología ---
df_tipologia = cargar_y_unificar(
    "../data/raw/accidents_segun_tipologia",
    ["Numero_expedient", "Codi_expedient"],
    ["Descripcio_tipus_accident", "Tipus_accident"]
)

df_causa_conductor = df_causa_conductor.rename(columns={'Descripcio_causa_mediata':'Descripcio_causa_conductor'})

# --- Merge final con df_accidents ---
df_accidents = df_accidents.merge(df_causa_conductor, on="Numero_expedient", how="left")
df_accidents = df_accidents.merge(df_causa_mediata, on="Numero_expedient", how="left")
df_accidents = df_accidents.merge(df_tipologia, on="Numero_expedient", how="left")
```

```python
# Rango de fechas
start_date = "2017-01-01"
end_date = "2024-12-31"

# Crear calendario de festivos para Barcelona (festivos nacionales, autonómicos y locales de Barcelona ciudad) - 14 por año
festivos_barcelona = holidays.country_holidays("ES", subdiv="CT", years=range(2017, 2024))

festivos_barcelona
```

```python
# Festivos locales de Barcelona: Segunda Pascua y La Mercè
segunda_pascua = {2017: "2017-06-05",
                  2018: "2018-05-21",
                  2019: "2019-06-10",
                  2020: "2020-06-01",
                  2021: "2021-05-24",
                  2022: "2022-06-06",
                  2023: "2023-05-29",
                  2024: "2024-05-20",}

la_merce = {2017: "2017-09-25",  # compensación
            2018: "2018-09-24",
            2019: "2019-09-24",
            2020: "2020-09-24",
            2021: "2021-09-24",
            2022: "2022-09-26",  # compensación
            2023: "2023-09-25",  # compensación
            2024: "2024-09-24",}
```

```python
# Añadir locales al calendario
for date_str in list(segunda_pascua.values()) + list(la_merce.values()):
    festivos_barcelona[pd.to_datetime(date_str)] = "Festivo local Barcelona"

len(festivos_barcelona)
```
